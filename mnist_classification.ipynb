```python
import tensorflow as tf
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Load in data
mnist = tf.keras.datasets.mnist

(x_train, y_train), (x_test, y_test) = mnist.load_data() # Split data into training and testing sets
```

```python
# Normalize features
x_train, x_test = x_train / 255.0, x_test / 255.0
```

```python
# Create Layers
model = tf.keras.models.Sequential([
    tf.keras.layers.Flatten(input_shape=(28, 28)),
    tf.keras.layers.Dense(128, activation='relu'),
    tf.keras.layers.Dropout(0.2),
    tf.keras.layers.Dense(10, activation='softmax')
])
```

```python
# Compile the model
model.compile (
    optimizer='adam',
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy']
)
```

```python
r = model.fit (
    x_train,
    y_train,
    validation_data=(x_test, y_test),
    epochs=10
)
```

```python
# Plot Accuracy
plt.plot(r.history['accuracy'], label='accuracy')
plt.plot(r.history['val_accuracy'], label='val_accuracy')
plt.legend()
```

```python
# Get all predictions
predictions = model.predict(x_test)
```

```python
# Predict the first 20 images
print('Predictions for the first 10 digits in the training set:')
for k in range(20):
  predicted_number = np.argmax(predictions[k])
  plt.imshow(x_test[k], cmap='gray')
  plt.title("Predicted Value: %s" % predicted_number)
  plt.show()
  print('\n')
```
